## Exercise 2 Solution 

````python
img = cv.imread("images/native-bee.png")
img2 = cv.imread("images/flower.jfif")

rows, cols, channels = img2.shape
roi = img[0:rows, 0:cols]

img2gray = cv.cvtColor(img2,cv.COLOR_BGR2GRAY)
ret, mask = cv.threshold(img2gray, 70, 255, cv.THRESH_BINARY) 
mask_inv = cv.bitwise_not(mask)

img_bg = cv.bitwise_and(roi,roi,mask = mask_inv)
img2_fg = cv.bitwise_and(img2,img2,mask = mask)

dst = cv.add(img_bg,img2_fg)
img[0:rows, 0:cols] = dst

display_image("result", img)
cv.waitKey(0)
cv.destroyAllWindows()
````

## Exercise 3 Solution

````python
def random_center_crop(image, min_crop_ratio, max_crop_ratio):
    height, width = image.shape[:2]   #0,1
    crop_ratio = random.uniform(min_crop_ratio, max_crop_ratio)
    crop_width, crop_height = int(width * crop_ratio), int(height * crop_ratio)
    mid_x, mid_y = width // 2, height // 2  #to get coordinates of original image
    start_x, start_y = mid_x - crop_width // 2, mid_y - crop_height // 2
    end_x, end_y = start_x + crop_width, start_y + crop_height
    return image[start_y:end_y, start_x:end_x]
    
````

````python
image = cv.imread("images/lena.jfif")
min_crop_ratio, max_crop_ratio = 0.5, 0.8
cropped_image = random_center_crop(image, min_crop_ratio, max_crop_ratio)
````

````python
# Display the original and cropped images
display_images([image,cropped_image],("Original","Cropped"))
````

## Exercise 4 Solution
Another common type of noise is :
1. Salt and pepper noise

````python
#Reference : https://medium.com/@Coursesteach/computer-vision-part-14-common-types-of-noise-7e6507cc763c

import matplotlib.pyplot as plt

def add_salt_and_pepper_noise(image, salt_prob, pepper_prob):
    noisy_image = np.copy(image)
    
    # Add salt noise
    salt_mask = np.random.rand(*image.shape) < salt_prob
    noisy_image[salt_mask] = 255

    # Add pepper noise
    pepper_mask = np.random.rand(*image.shape) < pepper_prob
    noisy_image[pepper_mask] = 0

    return noisy_image

# Load an example image
original_image = cv.imread('images/lena.jfif', cv.IMREAD_GRAYSCALE)

# Define the probabilities for salt and pepper noise
salt_probability = 0.01
pepper_probability = 0.01

# Add salt-and-pepper noise to the image
noisy_image = add_salt_and_pepper_noise(original_image, salt_probability, pepper_probability)

# Display the original and noisy images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(original_image, cv.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv.cvtColor(noisy_image, cv.COLOR_BGR2RGB))
plt.title('Noisy Image with Salt-and-Pepper Noise')
plt.axis('off')

plt.show()
````
